```python
client = pymongo.MongoClient()
db = client["tweet_database"]
tweet_table = db["tweet_table"]
```

```python
new_db = db.tweet_type_count

tweet_type = {'original': 0, 'retweets': 0}

for tweet in tweet_table.find():
    if tweet['favorited'] == True:
        tweet_type['original'] += 1
    else:
        tweet_type['retweets'] += 1

json = {"category": "Original Tweets", "Distribution": tweet_type['original']}
new_db.insert(json)
json = {"category": "Retweets", "Distribution": tweet_type['retweets']}
new_db.insert(json)
```

```python
hashtag_count = {}
new_db = db.hash_count
for tweet in tweet_table.find():
    ht = tweet["hashtags"]
    for hash in ht:
        hash = hash.lower()
        if hash not in hashtag_count:
            hashtag_count[hash] = 1
        else:
            hashtag_count[hash] += 1

hc = sorted(hashtag_count, key=hashtag_count.get, reverse=True)  # Sorting the top 10.
count = 0

for z in hc:
    if count == 10:
        break
    jsonx = {"hashtag": z, "count": str(hashtag_count[z])}
    new_db.insert(jsonx)
    count += 1
```
